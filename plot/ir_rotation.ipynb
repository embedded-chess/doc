```python
import numpy as np
import pandas as pd

df = pd.read_csv('angle_measurements.csv')

real = np.array(df.real)
meas = np.array(df.meas)
```

```python
import matplotlib.pyplot as plt

plt.rcParams['pgf.texsystem'] = 'pdflatex'
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 14.0
plt.rcParams['text.usetex'] = True
plt.rcParams['legend.loc'] = 'lower right'
plt.rcParams['savefig.bbox'] = 'tight'
plt.rcParams['figure.figsize'] = (8, 5)


plt.plot(real, meas, 'blue')
plt.plot(real, real, color='grey', linestyle='--')

plt.xlabel(r'Tats√§chlicher Winkel $\alpha_{r}$')
plt.ylabel(r'Gemessener Winkel $\alpha_{m}$')

plt.xlim(0, 360)
plt.ylim(0, 360)

plt.savefig('ir_signal_comparison.pdf', format='pdf')
plt.show()
```

```python
plt.plot(real, meas - real, color='black', linewidth=1)
plt.scatter(real, meas - real)

plt.grid(which='both', alpha=0.5)

plt.xticks(np.linspace(0, 360, 13))
plt.axhline(0, color='grey')

plt.xlabel(r'Tats√§chlicher Winkel $\alpha_{r}$')
plt.ylabel(r'Differenz der Messung $\Delta\alpha$')

plt.xlim(-10, 360)
plt.ylim(-11, 11)

plt.savefig('ir_signal_diff.pdf', format='pdf')
plt.show()
```
